- [Albert Einstein](/wiki/Albert_Einstein) suggested a [thought experiment](/wiki/Einstein%27s_thought_experiments) involving two objects that interact and then move apart.
- The quantum-mechanical description of the two objects is a mathematical entity known as a [wavefunction](/wiki/wavefunction).
- If the wavefunction that describes the two objects before their interaction is given, then the [Schrödinger equation](/wiki/Schr%C3%B6dinger_equation) provides the wavefunction that describes them after their interaction.
- The phenomenon known as [quantum entanglement](/wiki/quantum_entanglement) causes measuring one object to lead to an instantaneous change of the wavefunction describing the other object.
- The instantaneous change in the wavefunction occurs regardless of how far apart the two objects are.
- The choice of which measurement to perform upon the first object affects what wavefunction could result for the second object.
- [Albert Einstein](/wiki/Albert_Einstein) reasoned that no influence could propagate from the first object to the second instantaneously fast.
- [Albert Einstein](/wiki/Albert_Einstein) argued that physics depends on being able to tell one thing apart from another.
- [Albert Einstein](/wiki/Albert_Einstein) claimed that instantaneous influences would call this into question.
- Because the true "physical condition" of the second object could not be immediately altered by an action done to the first, [Albert Einstein](/wiki/Albert_Einstein) concluded that the wavefunction could not be that true physical condition.
- [Albert Einstein](/wiki/Albert_Einstein) believed the wavefunction was only an incomplete description of the true physical condition.

- [Albert Einstein](/wiki/Albert_Einstein) published a paper in 1935 with [Boris Podolsky](/wiki/Boris_Podolsky) and [Nathan Rosen](/wiki/Nathan_Rosen).
- The paper published by [Albert Einstein](/wiki/Albert_Einstein), [Boris Podolsky](/wiki/Boris_Podolsky), and [Nathan Rosen](/wiki/Nathan_Rosen) in 1935 laid out what would become known as the [EPR paradox](/wiki/EPR_paradox).
- In the [EPR paradox](/wiki/EPR_paradox), two particles interact in such a way that the wavefunction describing them is entangled.
- In the [EPR paradox](/wiki/EPR_paradox), no matter how far the two particles were separated, a precise position measurement on one particle would imply the ability to predict, perfectly, the result of measuring the position of the other particle.
- In the [EPR paradox](/wiki/EPR_paradox), a precise momentum measurement of one particle would result in an equally precise prediction for the momentum of the other particle.
- In the [EPR paradox](/wiki/EPR_paradox), no disturbance of the other particle was needed to make the prediction about its momentum.
- [Albert Einstein](/wiki/Albert_Einstein), [Boris Podolsky](/wiki/Boris_Podolsky), and [Nathan Rosen](/wiki/Nathan_Rosen) argued that no action taken on the first particle could instantaneously affect the other.
- [Albert Einstein](/wiki/Albert_Einstein), [Boris Podolsky](/wiki/Boris_Podolsky), and [Nathan Rosen](/wiki/Nathan_Rosen) claimed that information transmission faster than light is forbidden by the [theory of relativity](/wiki/theory_of_relativity).
- The [EPR paradox](/wiki/EPR_paradox) invoked a principle later known as the "EPR criterion of reality".
- The "EPR criterion of reality" posits that if, without in any way disturbing a system, we can predict with certainty the value of a physical quantity, then there exists an element of reality corresponding to that quantity.
- From the "EPR criterion of reality", [Albert Einstein](/wiki/Albert_Einstein), [Boris Podolsky](/wiki/Boris_Podolsky), and [Nathan Rosen](/wiki/Nathan_Rosen) inferred that the second particle must have a definite value of both position and momentum prior to either quantity being measured.
- Quantum mechanics considers position and momentum observables [incompatible](/wiki/Observable).
- Quantum mechanics does not associate simultaneous values for both position and momentum to any system.
- [Albert Einstein](/wiki/Albert_Einstein), [Boris Podolsky](/wiki/Boris_Podolsky), and [Nathan Rosen](/wiki/Nathan_Rosen) concluded that quantum theory does not provide a complete description of reality.

- In 1964, [John Stewart Bell](/wiki/John_Stewart_Bell) carried the analysis of quantum entanglement much further.
- [John Stewart Bell](/wiki/John_Stewart_Bell) deduced that if measurements are performed independently on the two separated particles of an entangled pair, then the assumption that the outcomes depend upon hidden variables within each half implies a mathematical constraint on how the outcomes on the two measurements are correlated.
- This constraint would later be called a [Bell inequality](/wiki/Bell_inequality).
- [John Stewart Bell](/wiki/John_Stewart_Bell) showed that quantum physics predicts correlations that violate this inequality.
- The only way that hidden variables could explain the predictions of quantum physics is if they are "nonlocal".
- Nonlocality implies that the two particles are able to interact instantaneously no matter how widely they ever become separated.
- [John Stewart Bell](/wiki/John_Stewart_Bell) argued that because an explanation of quantum phenomena in terms of hidden variables would require nonlocality, the [EPR paradox](/wiki/EPR_paradox) "is resolved in the way which Einstein would have liked least".
- Despite this, [Albert Einstein](/wiki/Albert_Einstein) personally found the argument in the [EPR paper](/wiki/EPR_paradox) overly complicated.
- The [EPR paper](/wiki/EPR_paradox) became among the most influential papers published in *[Physical Review](/wiki/Physical_Review)*.
- The [EPR paper](/wiki/EPR_paradox) is considered a centerpiece of the development of [quantum information theory](/wiki/quantum_information_theory).
