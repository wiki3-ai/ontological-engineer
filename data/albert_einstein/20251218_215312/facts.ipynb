• No matter how far apart the two entangled particles were separated, a precise position measurement on one particle would allow perfect prediction of the position of the other particle.
• No matter how far apart the two entangled particles were separated, a precise momentum measurement on one particle would allow perfect prediction of the momentum of the other particle.
• Einstein, Podolsky, and Rosen argued that no action taken on the first particle could instantaneously affect the other particle because this would involve information being transmitted faster than light.
• The EPR paradox invoked a principle later known as the "EPR criterion of reality" which states that if we can predict with certainty the value of a physical quantity without disturbing a system, then there exists an element of reality corresponding to that quantity.
• From the EPR criterion of reality, Einstein, Podolsky, and Rosen inferred that the second particle must have a definite value of both position and momentum prior to either quantity being measured.
• Quantum mechanics considers position and momentum observables to be incompatible and does not associate simultaneous values for both to any system.
• Einstein, Podolsky, and Rosen concluded that quantum theory does not provide a complete description of reality.

• John Stewart Bell carried the analysis of quantum entanglement much further in 1964
• Bell deduced that if measurements are performed independently on two separated particles of an entangled pair, then the assumption that outcomes depend upon hidden variables within each half implies a mathematical constraint on measurement correlations
• This constraint later became known as a Bell inequality
• Bell showed that quantum physics predicts correlations that violate the Bell inequality
• Bell argued that hidden variables could only explain quantum physics predictions if they were "nonlocal"
• Bell stated that nonlocality would require the two particles to interact instantaneously regardless of separation distance
• Bell argued that because hidden variables would require nonlocality, the EPR paradox "is resolved in the way which Einstein would have liked least"
• Although Einstein personally found the argument in the EPR paper overly complicated, that paper became among the most influential papers published in Physical Review
• The EPR paper is considered a centerpiece of the development of quantum information theory

- Owen Willans Richardson predicted in 1908 that a change in the magnetic moment of a free body would cause it to rotate.
- The effect predicted by Richardson is a consequence of the conservation of angular momentum.
- The effect is strong enough to be observable in ferromagnetic materials.
- Albert Einstein and Wander Johannes de Haas published two papers in 1915 claiming the first experimental observation of the Einstein-de Haas effect.
- The Einstein-de Haas experiment demonstrates that magnetization is caused by the alignment of the angular momenta of electrons in a material.
- The experiment allows the separation of contributions to magnetization from electron spin and orbital motion.
- The Einstein-de Haas experiment was the only experiment conceived, realized, and published by Albert Einstein himself.
- A complete original version of the Einstein-de Haas experimental equipment was donated to the Ampère Museum in Lyon, France, in 1961.
- The Einstein-de Haas experimental equipment was lost among the museum's holdings and was rediscovered in 2023.
- In 1926, Albert Einstein and his former student Leó Szilárd co-invented the refrigerator known as the Einstein refrigerator.
- The Einstein refrigerator was patented in 1930 under U.S. Patent No. 1,781,541.
- The refrigerator was revolutionary for having no moving parts and using only heat as an input.
- The invention was not immediately put into commercial production, but some of its patents were acquired by Electrolux.
- Albert Einstein also invented an electromagnetic pump, a sound device, and several other household devices.
