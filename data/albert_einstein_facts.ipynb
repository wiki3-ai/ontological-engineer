• The EPR paradox demonstrated that a precise momentum measurement on one particle could predict the momentum of the other particle perfectly.
• Einstein, Podolsky, and Rosen argued that no action taken on the first particle could instantaneously affect the other particle.
• Einstein, Podolsky, and Rosen believed that information transmission faster than light would violate the theory of relativity.
• Einstein, Podolsky, and Rosen invoked a principle later known as the "EPR criterion of reality."
• The EPR criterion of reality states that if we can predict with certainty the value of a physical quantity without disturbing a system, then there exists an element of reality corresponding to that quantity.
• Einstein, Podolsky, and Rosen inferred that the second particle must have definite values of both position and momentum before measurement.
• Quantum mechanics considers position and momentum observables to be incompatible and does not associate simultaneous values for both to any system.
• Einstein, Podolsky, and Rosen concluded that quantum theory does not provide a complete description of reality.

- Owen Willans Richardson predicted in 1908 that a change in the magnetic moment of a free body would cause it to rotate.
- The effect predicted by Richardson is a consequence of the conservation of angular momentum.
- The effect is strong enough to be observable in ferromagnetic materials.
- Albert Einstein and Wander Johannes de Haas published two papers in 1915 claiming the first experimental observation of the Einstein-de Haas effect.
- The Einstein-de Haas experiment demonstrates that magnetization is caused by the alignment of the angular momenta of electrons in a material.
- The experiment allows the separation of contributions to magnetization from electron spin and orbital motion.
- The Einstein-de Haas experiment was the only experiment conceived, realized, and published by Albert Einstein himself.
- A complete original version of the Einstein-de Haas experimental equipment was donated to the Ampère Museum in Lyon, France, in 1961.
- The equipment was lost among the museum's holdings and was rediscovered in 2023.
- In 1926, Albert Einstein and his former student Leó Szilárd co-invented the refrigerator known as the Einstein refrigerator.
- The Einstein refrigerator was patented in 1930 under U.S. Patent No. 1,781,541.
- The refrigerator used heat as input and had no moving parts.
- Einstein and Szilárd's invention was not immediately put into commercial production.
- The Swedish company Electrolux acquired the most promising of their patents.
- Albert Einstein also invented an electromagnetic pump, a sound device, and several other household devices.
