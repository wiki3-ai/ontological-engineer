**Context:** Albert Einstein > Scientific career > Quantum mechanics > Einstein–Podolsky–Rosen paradox

---

• Albert Einstein published a paper in 1935 with Boris Podolsky and Nathan Rosen that laid out what became known as the EPR paradox.
• The EPR paradox involved a thought experiment with two particles that interacted in such a way that their wavefunction was entangled.
• The EPR paradox showed that measuring one particle could allow perfect prediction of the other particle's position or momentum.
• The EPR paradox argued that no action on the first particle could instantaneously affect the second particle because this would involve faster-than-light information transmission.
• Einstein, Podolsky, and Rosen invoked a principle later known as the "EPR criterion of reality" stating that if we can predict with certainty the value of a physical quantity without disturbing the system, then there exists an element of reality corresponding to that quantity.
• The EPR paradox concluded that the second particle must have definite values of both position and momentum before measurement.
• Quantum mechanics considers position and momentum to be incompatible observables that cannot have simultaneous values for any system.
• Einstein, Podolsky, and Rosen concluded that quantum theory does not provide a complete description of reality.
