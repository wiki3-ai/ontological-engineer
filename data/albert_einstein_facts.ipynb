• Quantum entanglement causes measuring one object to lead to an instantaneous change of the wavefunction describing the other object.
• The choice of measurement performed on the first object affects what wavefunction could result for the second object.
• Einstein reasoned that no influence could propagate from the first object to the second instantaneously fast.
• Einstein argued that physics depends on being able to tell one thing apart from another.
• Einstein claimed instantaneous influences would call this principle into question.
• Einstein concluded that the wavefunction could not be the true physical condition of an object.
• Einstein believed the wavefunction was only an incomplete description of the true physical condition.


**Context:** Albert Einstein > Scientific career > Quantum mechanics > Einstein–Podolsky–Rosen paradox
**Chunk:** 55 of 63

---

• Albert Einstein published a paper with Boris Podolsky and Nathan Rosen in 1935 that laid out the EPR paradox.
• The EPR paradox involved a thought experiment with two entangled particles that interacted in such a way that their wavefunction was entangled.
• The EPR paradox showed that measuring one particle could allow perfect prediction of the other particle's position or momentum.
• The EPR paradox argued that no action on the first particle could instantaneously affect the second particle because this would involve faster-than-light information transmission.
• Einstein, Podolsky, and Rosen invoked a principle known as the "EPR criterion of reality" stating that if we can predict with certainty the value of a physical quantity without disturbing the system, then there exists an element of reality corresponding to that quantity.
• The EPR paradox concluded that the second particle must have definite values of both position and momentum before measurement.
• Quantum mechanics considers position and momentum observables to be incompatible and does not associate simultaneous values for both to any system.
• Einstein, Podolsky, and Rosen concluded that quantum theory does not provide a complete description of reality.

• John Stewart Bell carried the analysis of quantum entanglement further in 1964
• Bell deduced that if measurements are performed independently on two separated particles of an entangled pair, then the assumption that outcomes depend upon hidden variables within each half implies a mathematical constraint on how the outcomes are correlated
• This constraint later became known as a Bell inequality
• Bell showed that quantum physics predicts correlations that violate this inequality
• The only way hidden variables could explain the predictions of quantum physics is if they are "nonlocal"
• Nonlocality means that the two particles are able to interact instantaneously no matter how widely they become separated
• Bell argued that because an explanation of quantum phenomena in terms of hidden variables would require nonlocality, the EPR paradox is resolved in the way which Einstein would have liked least
• Although Einstein personally found the argument in the EPR paper overly complicated, that paper became among the most influential papers published in Physical Review
• The EPR paper is considered a centerpiece of the development of quantum information theory

- Owen Willans Richardson predicted in 1908 that a change in the magnetic moment of a free body would cause it to rotate.
- The effect predicted by Richardson is a consequence of the conservation of angular momentum.
- The effect is strong enough to be observable in ferromagnetic materials.
- Albert Einstein and Wander Johannes de Haas published two papers in 1915 claiming the first experimental observation of the Einstein–de Haas effect.
- The Einstein–de Haas experiment demonstrates that magnetization is caused by the alignment of electron angular momenta in a material.
- The experiment allows separation of contributions to magnetization from electron spin and orbital motion.
- The Einstein–de Haas experiment was the only experiment conceived, realized, and published by Albert Einstein himself.
- A complete original version of the Einstein–de Haas experimental equipment was donated to the Ampère Museum in Lyon, France, in 1961.
- The equipment was lost among the museum's holdings and was rediscovered in 2023.
- In 1926, Albert Einstein and his former student Leó Szilárd co-invented the refrigerator known as the Einstein refrigerator.
- The Einstein refrigerator was patented in 1930 under U.S. Patent No. 1,781,541.
- The refrigerator used heat as its input and had no moving parts.
- Einstein and Szilárd's invention was not immediately put into commercial production.
- The Swedish company Electrolux acquired the most promising of their patents.
- Albert Einstein also invented an electromagnetic pump, a sound device, and several other household devices.
